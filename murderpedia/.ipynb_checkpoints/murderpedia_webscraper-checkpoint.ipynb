```python
import pandas as pd
import os
import codecs
import time
import requests
import re
from urllib.request import Request, urlopen, build_opener
from bs4 import BeautifulSoup
```

```python
start_url = 'https://murderpedia.org/'
req = Request(start_url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.84 Safari/537.36'})
page = urlopen(req).read()
soup = BeautifulSoup(page,'html.parser')#str(BeautifulSoup(page,'html.parser'))
print(soup)
```

```python
##Pull full list of root pages
tmp = soup.findAll('a',  {"href" : re.compile('male.*')}) # finds all <a> tags
all_roots = [];
for it in range(len(tmp)):
    if tmp[it]['href'].find('/') < 0:
        continue
    else:
        all_roots += [start_url+tmp[it]['href']]
        
df = pd.DataFrame(all_roots, columns = ['URL'])
df
```

```python
test_url = all_roots[0]
req2 = Request(test_url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.84 Safari/537.36'})
page2 = urlopen(req2).read()
soup2 = BeautifulSoup(page2,'html.parser')
#pull everything in tbody
soup2.findAll('a',{"href": re.compile('(.htm)')})

```
